```python
from sentence_transformers import SentenceTransformer, losses, InputExample
from torch.utils.data import DataLoader
```

```python
# load data
#data = pd.read_csv('fashion_len.csv')
data = pd.read_csv('fashion_len.csv')
```

```python
data
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
# run data in batches due to memory limitations
batch_size = 100
for e in range(math.ceil(len(data)/batch_size)):
    start = e*batch_size
    end = min((e+1)*batch_size, len(data))
    batch = data['text'][start: end]
    # construct feed to dataloader
    train_feed = [InputExample(texts = list(batch), label = 0.0)]
    # load data to pytorch dataloader
    train_dataloader = DataLoader(train_feed, shuffle=True, batch_size=8)
    # set train loss to cosine similarity
    train_loss = losses.CosineSimilarityLoss(model)
    model.fit(train_objectives = [(train_dataloader, train_loss)], epochs = 1, warmup_steps = 0)
```

```python
# load data
data = pd.read_csv('fashion_len.csv')
```

```python
# combine embeddings and star ratings
train = np.column_stack((train_embedded, np.array(data['star'])))
```

```python
def cluster_sentences(number_clusters, data, max_iter = 50):
  minikmedoids = MiniBatchKMedoids(n_clusters=number_clusters, max_iter=max_iter, batch_size=100, metric='cosine', 
                                   max_no_improvement=10, random_state=42)
  minikmedoids.fit([data])
  clusters = minikmedoids.cluster_centers_
  labels = minikmedoids.predict([data])[0]

  #get error
  error = 0
  sentences = []
  
  for i in set(labels):
    cluster_items = data[np.where(labels==i)]
    min_dist = np.inf
    min_item = None
    cluster_center = clusters[i]
    for c in cluster_items:
      dist = distance.cosine(c, cluster_center)
      if dist < min_dist:
        min_dist = dist
        min_item = c
      error += dist
    sentences.append(min_item)
  return minikmedoids, clusters, error, sentences, labels
```

```python
errors = []
clusters_range = np.arange(5,30,5) # run against different numbers of cluster initializations
df_clusters = pd.DataFrame(columns=['n_nominal','n_real', 'clusters', 'e', 'labels'])
dict_cluster = {}
for i, c in enumerate(clusters_range):
  #print(c)
  model, clusters, e, sentences, labels = cluster_sentences(c, train)
  #print(sentences)
  unique_labels = len(set(labels))
  df_clusters.loc[i, 'n_nominal'] = c
  df_clusters.loc[i, 'n_real'] = unique_labels
  df_clusters.loc[i, 'clusters'] = clusters
  df_clusters.loc[i, 'e'] = e
  df_clusters.loc[i, 'labels'] = labels
  dict_cluster[c] ={}
  dict_cluster[c]['n_real'] = unique_labels
  dict_cluster[c]['sentences'] = sentences
  dict_cluster[c]['model'] = model
```

```python
df_clusters
```

```python
# visualize cluster error - not trained
x = df_clusters['n_real'].astype(int)
y = df_clusters['e']
z = df_clusters['n_nominal'].astype(int)
plt.scatter(x, y)
plt.xlabel('Number of Clusters')
plt.ylabel('Cluster Error')
plt.xlim(4, 10)
plt.ylim(2870, 2905)
plt.xticks(x)

# annotate
k = 0
for xi,yj in zip(x,y):

    label = "{:.2f}".format(yj)
    if k==1:
      plt.annotate("init = " + str(z[k]), # this is the text
                  (xi,yj), # these are the coordinates to position the label
                  textcoords="offset points", # how to position the text
                  xytext=(0,-15), # distance from text to points (x,y)
                  ha='center') # horizontal alignment can be left, right or center
    else:
      plt.annotate("init = " + str(z[k]), # this is the text
                 (xi,yj), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
    k +=1
plt.show()
```
